### A summary of some important conceptual points

<center></center>

**Some important caveats to keep keep in mind about amplicon sequencing:**
- "universal" primers try to capture a large breadth of diversity, and they do a good job, but they of course don't capture all variations of our target gene that exist
- many ecologically relevant biological entities, like viruses and plasmids, are typically not captured at all by this approach
- does not really provide any information on functional potential
- when using a commonly used target gene like the SSU rRNA (small-subunit ribosomal RNA; 16S in bacteria/archaea, 18S in eukarya), **the counts we get from processing amplicon data represent counts of recovered gene-copies; they do not represent counts of organisms or genomes**
  - a few reasons for this include:
    - gene-copy number present per organism/genome can vary as much as from 1 to greater than 20 (and they are not necessarily identical within a genome)
    - amplification bias (small scale)
      - the primers used have different binding efficiencies for different sequences and they will inherently over-amplify some sequences while under-amplifying others
    - amplification bias (large scale)
      - like mentioned above, "universal" primers are pretty good in the breadth of diversity they can capture, but they still can be missing microbes from any given sample
    - cell-lysis efficiencies
      - different cells break apart at different rates (and some maybe not at all), so right from the start we are introducing biases in the DNA we might potentially amplify and sequence
  - **main takeaway** is that the counts of gene-copies we get from processing amplicon data do not equal organism or genome abundances
    - however, this approach can still be useful as a metric of what microbial communities look like, and seeing differences in counts of recovered gene-copies can still reflect something real about our system and help guide hypotheses 👍
